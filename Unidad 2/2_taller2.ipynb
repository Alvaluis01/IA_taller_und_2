Su tarea es programar la lógica de un agente para encontrar la salida de un _laberinto_. Para facilitar el ejercicio, se le proporciona una clase ``Laberinto`` que gestiona el mapa y los movimientos. Usted debe enfocarse en implementar el algoritmo de búsqueda.

```python
# Clase con los métodos necesarios para transitar el laberinto
class Laberinto:
    def __init__(self, mapa):
        self.mapa = mapa
        self.alto = len(mapa)
        self.ancho = len(mapa[0])
        self.inicio = self._encontrar_caracter('S')
        self.fin = self._encontrar_caracter('E')

    def _encontrar_caracter(self, caracter):
        """Encuentra las coordenadas (fila, col) de un caracter en el mapa."""
        for f, fila in enumerate(self.mapa):
            try:
                c = fila.index(caracter)
                return (f, c)
            except ValueError:
                continue
        return None

    def es_meta(self, posicion):
        """Verifica si una posición es la meta."""
        return posicion == self.fin

    def obtener_movimientos_validos(self, posicion):
        """Devuelve una lista de posiciones vecinas transitables."""
        fila, col = posicion
        movimientos = []
        candidatos = [(fila - 1, col), (fila + 1, col), 
                      (fila,col - 1), (fila, col + 1)] # Arriba, Abajo, Izq, Der

        for f, c in candidatos:
            if 0 <= f < self.alto and 0 <= c < self.ancho and self.mapa[f][c] != '#':
                movimientos.append((f, c))
        return movimientos
```
